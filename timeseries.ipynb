## Introduction to Time Series Data

Time series data is a sequence of data points collected or recorded over regular time intervals. It can be found in various fields, such as finance, meteorology, and health care, and is used to study the evolution of a variable over time. Examples of time series data include stock prices, temperature readings, and daily COVID-19 cases.

Time series analysis aims to understand the underlying patterns in the data, such as trends, seasonality, and noise. It also seeks to develop models that can forecast future values based on historical data. Forecasting is crucial for decision-making in various domains, such as business planning, resource allocation, and policy formulation.

## Time Series Data Considerations
When analyzing time series data, it is essential to consider:

* Trend: The overall direction of the data (e.g., increasing, decreasing, or constant)

* Seasonality: The presence of recurring patterns or cycles in the data (e.g., daily, weekly, or yearly)

* Noise: The random fluctuations in the data that are not explained by the trend or seasonality

* Autocorrelation: The correlation between a time series and its lagged values (i.e., its past values)

* Stationarity: A time series is stationary if its statistical properties (mean, variance, and autocorrelation) do not change over time. 
Most time series models assume stationarity, so it is essential to test for and, if necessary, transform the data to achieve stationarity.

Time Series Data Considerations
------------------------------------------

When analyzing time series data, it is essential to consider several factors that could impact the structure and behavior of the data. These factors include:

1. **Trend**: The overall direction of the data over time (e.g., increasing, decreasing, or constant). Trends can be linear or nonlinear and can change over time. Identifying and accounting for trends is crucial in time series analysis, as it can help separate the underlying structure from the noise.

2. **Seasonality**: The presence of recurring patterns or cycles in the data that occur at regular intervals (e.g., daily, weekly, or yearly). Seasonality can be additive or multiplicative and is often related to external factors, such as weather or holidays. Accounting for seasonality can help improve the accuracy of forecasts.

3. **Noise**: The random fluctuations in the data that are not explained by the trend or seasonality. Noise can result from measurement errors, random events, or other unexplained sources of variation. Noise can make it challenging to identify the underlying structure in the data, so it is essential to differentiate it from the signal (i.e., the trend and seasonality).

4. **Autocorrelation**: The correlation between a time series and its lagged values (i.e., its past values). Autocorrelation can be positive or negative and can help identify the presence of trends, seasonality, or other patterns in the data. Autocorrelation plots can be used to visualize the relationship between a time series and its lagged values.

5. **Stationarity**: A time series is stationary if its statistical properties (mean, variance, and autocorrelation) do not change over time. Most time series models assume stationarity, so it is essential to test for and, if necessary, transform the data to achieve stationarity. Techniques to test for stationarity include the Augmented Dickey-Fuller (ADF) test and the Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test.



## Time Series Visualization
Visualizing time series data is an essential step in understanding its structure and behavior. Visualization can help identify trends, seasonality, and noise in the data. In this section, we will explore various techniques for visualizing time series data, including line plots, autocorrelation plots, and partial autocorrelation plots.

**Line Plots**

Line plots are the most common way to visualize time series data. They display the data points connected by lines, showing the evolution of the variable over time.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load time series data
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, parse_dates=True, index_col='Month')

# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data)
plt.xlabel('Time')
plt.ylabel('Passengers (Thousands)')
plt.title('Monthly U.S. Airline Passenger Numbers')
plt.show()

````

**Autocorrelation Plots**

Autocorrelation plots, also known as correlograms, display the correlation between a time series and its lagged values. These plots can help identify trends and seasonality in the data.

````python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Plot the autocorrelation
plot_acf(data)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Plot')
plt.show()
````

**Partial Autocorrelation Plots**

Partial autocorrelation plots display the correlation between a time series and its lagged values after accounting for the correlations of the intervening lags. These plots can help identify the order of autoregressive models.

````python
from statsmodels.graphics.tsaplots import plot_pacf
# Plot the partial autocorrelation
plot_pacf(data)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Plot')
plt.show()
````

## Time Series Decomposition

Time series decomposition is a technique for breaking down a time series into its individual components, such as trend, seasonality, and noise. Decomposition can help improve our understanding of the data and inform the choice of models and preprocessing techniques used in time series analysis. In this section, we will explore additive and multiplicative decomposition using the seasonal_decompose function from the statsmodels library.

**Additive Decomposition**

Additive decomposition assumes that the time series can be represented as the sum of its components:

Time Series = Trend + Seasonality + Noise

````python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform additive decomposition
additive_decomposition = seasonal_decompose(data, model='additive')

# Plot the additive decomposition
additive_decomposition.plot()
plt.show()
````

## Multiplicative Decomposition

Multiplicative decomposition assumes that the time series can be represented as the product of its components:

Time Series = Trend * Seasonality * Noise

````python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform multiplicative decomposition
multiplicative_decomposition = seasonal_decompose(data, model='multiplicative')

# Plot the multiplicative decomposition
multiplicative_decomposition.plot()
plt.show()
````

## Stationarity and Differencing

Stationarity is an important property of time series data that affects the performance of time series models. A stationary time series has constant mean, variance, and autocorrelation over time. Many time series models, such as ARIMA, assume that the data is stationary. If the data is not stationary, it can be transformed using differencing to make it stationary.

Differencing is a technique for making a non-stationary time series stationary by subtracting the previous value from the current value. The order of differencing (d) determines the number of times the differencing operation is applied to the original time series. The first-order differenced series is obtained by:

Y'_t = Y_t - Y_(t-1)

where Y_t is the value at time t, and Y'_t is the differenced value at time t.


````python
import pandas as pd
import matplotlib.pyplot as plt

# First-order differencing
data_diff = data.diff().dropna()

# Plot the original and differenced time series
fig, ax = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

ax[0].plot(data)
ax[0].set_title('Original Time Series')
ax[0].set_ylabel('Passengers (Thousands)')

ax[1].plot(data_diff)
ax[1].set_title('First-Order Differenced Time Series')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Differenced Passengers')

plt.show()
````

The graph displays two subplots that show the original time series data and the first-order differenced time series data.

1. The first subplot (Original Time Series) shows the "Monthly U.S. Airline Passenger Numbers" from January 1949 to December 1960. The x-axis represents time, and the y-axis represents the number of passengers (in thousands). This plot helps visualize the overall trend and seasonality present in the data.

2. The second subplot (First-Order Differenced Time Series) shows the first-order differenced time series obtained by subtracting the previous month's passenger number from the current month's passenger number. The x-axis represents time, and the y-axis represents the difference in passengers (in thousands) between consecutive months. This plot helps visualize the changes in the time series after applying the differencing operation, which can make a non-stationary time series stationary.

By comparing the two subplots, we can observe how the differencing operation has transformed the original time series, potentially making it suitable for time series models that assume stationarity, such as ARIMA.

## Section 5: Time Series Modeling

In this section, we will explore various time series models and their assumptions, pitfalls, and appropriate uses. We will cover autoregressive (AR) models, moving average (MA) models, autoregressive integrated moving average (ARIMA) models, and seasonal decomposition of time series (STL) models.

### 5.1 Autoregressive (AR) Models

Autoregressive models assume that the current value of a time series depends linearly on its previous values and a random error term. AR models can be represented as:

`Y_t = α + β_1 * Y_(t-1) + β_2 * Y_(t-2) + ... + β_p * Y_(t-p) + ε_t`

where `Y_t` is the value at time t, α is a constant term, β_i are the coefficients of the model, p is the order of the model, and ε_t is the error term.

#### 5.1.1 AR Model Assumptions

1. The time series is stationary.
2. The relationship between the current value and its previous values is linear.

#### 5.1.2 AR Model Pitfalls

1. The model may not capture complex, non-linear relationships in the data.
2. If the true underlying process is not autoregressive, the model may produce inaccurate forecasts.

#### 5.1.3 Appropriate Uses for AR Models

1. When the time series is stationary, and there is a linear relationship between the current value and its previous values.
2. When the true underlying process is autoregressive.

### 5.2 Moving Average (MA) Models

Moving average models assume that the current value of a time series is a linear combination of past error terms. MA models can be represented as:

`Y_t = μ + ε_t + θ_1 * ε_(t-1) + θ_2 * ε_(t-2) + ... + θ_q * ε_(t-q)`

where `Y_t` is the value at time t, μ is the mean of the series, ε_t is the error term, θ_i are the coefficients of the model, and q is the order of the model.

#### 5.2.1 MA Model Assumptions

1. The time series is stationary.
2. The relationship between the current value and past error terms is linear.

#### 5.2.2 MA Model Pitfalls

1. The model may not capture complex, non-linear relationships in the data.
2. If the true underlying process is not a moving average process, the model may produce inaccurate forecasts.

#### 5.2.3 Appropriate Uses for MA Models

1. When the time series is stationary, and there is a linear relationship between the current value and past error terms.
2. When the true underlying process is a moving average process.

### 5.3 Autoregressive Integrated Moving Average (ARIMA) Models

ARIMA models combine autoregressive and moving average models and use differencing to make a non-stationary time series stationary. ARIMA models can be represented as:

`Y'_t = α + β_1 * Y'_(t-1) + ... + β_p * Y'_(t-p) + ε_t + θ_1 * ε_(t-1) + ... + θ_q * ε_(t-q)`

where `Y'_t` is the differenced value at time t, α is a constant term, β_i and θ_i are the coefficients of the model, p is the order of the autoregressive part, q is the order of the moving average part, and ε_t is the error term. The model is denoted as ARIMA(p, d, q), where d is the order of differencing.

#### 5.3.1 ARIMA Model Assumptions

1. The time series is stationary after differencing (d times).
2. The relationship between the current differenced value, its previous differenced values, and past error terms is linear.

#### 5.3.2 ARIMA Model Pitfalls

1. The model may not capture complex, non-linear relationships in the data.
2. If the true underlying process is not an autoregressive or moving average process, the model may produce inaccurate forecasts.
3. The process of selecting the best combination of p, d, and q can be time-consuming and computationally expensive.

#### 5.3.3 Appropriate Uses for ARIMA Models

1. When the time series becomes stationary after differencing, and there is a linear relationship between the current differenced value, its previous differenced values, and past error terms.
2. When the true underlying process is an autoregressive or moving average process.

### 5.4 Seasonal Decomposition of Time Series (STL) Models

STL models decompose a time series into its trend, seasonal, and residual components. This decomposition can be used as a preprocessing step for time series modeling. The seasonal decomposition can be additive or multiplicative.

#### 5.4.1 STL Model Assumptions

1. The time series can be decomposed into trend, seasonal, and residual components.

#### 5.4.2 STL Model Pitfalls

1. The model may not be appropriate for time series with rapidly changing seasonality or non-linear trends.
2. If the true underlying process does not have a clear seasonal component, the model may not provide useful information.

#### 5.4.3 Appropriate Uses for STL Models

1. When the time series exhibits clear seasonal patterns and can be decomposed into trend, seasonal, and residual components.
2. As a preprocessing step to remove seasonality before applying other time series models.

In this section, we have explored various time series models, their assumptions, pitfalls, and appropriate uses. Understanding these models and their properties is essential for selecting the best model for a given time series and obtaining accurate forecasts.

## Section 6: Model Selection and Evaluation

In this section, we will discuss how to select the best time series model for a given dataset and evaluate its performance using various metrics.

### 6.1 Model Selection

Model selection involves choosing the best time series model based on the characteristics of the data and the assumptions of the models. Here are some general guidelines for selecting a model:

1. Examine the time series data and perform exploratory data analysis to identify trends, seasonality, and any other patterns.
2. Check if the data is stationary or if differencing is required. Apply appropriate transformations if necessary.
3. Test various models, such as AR, MA, ARIMA, and STL, based on the data characteristics and model assumptions.
4. Use techniques such as cross-validation, grid search, or information criteria (e.g., Akaike Information Criterion (AIC) or Bayesian Information Criterion (BIC)) to select the best model with optimal parameters.

### 6.2 Model Evaluation

Model evaluation involves assessing the performance of a time series model using various metrics. Some common evaluation metrics for time series models include:

1. Mean Absolute Error (MAE): The average of the absolute errors between the predicted values and the actual values.
2. Mean Squared Error (MSE): The average of the squared errors between the predicted values and the actual values.
3. Root Mean Squared Error (RMSE): The square root of the MSE, which has the same units as the original data.
4. Mean Absolute Percentage Error (MAPE): The average of the absolute percentage errors between the predicted values and the actual values.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

y_true = np.array([10, 20, 30, 40, 50])
y_pred = np.array([12, 18, 29, 39, 52])

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_true, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
```

In this section, we have discussed model selection and evaluation for time series models. Proper model selection and evaluation are crucial for obtaining accurate forecasts and assessing the performance of the chosen model.
